## Merging Crash and Violations Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('crashdata.csv')
df2 = pd.read_csv('Speed_Camera_Violations_merge.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3 = df2.groupby('CAMERA ID')['VIOLATIONS'].sum().reset_index()
```

```python
df3.head()
```

```python
df_11 = df1.copy()
```

```python
df_11[['camID1','camID2']] = df_11.Camera_ID.str.split("|",expand=True)
```

```python
df_11['camID1'] = df_11['camID1'].str.strip('*')
df_11['camID2'] = df_11['camID2'].str.strip('*')
```

```python
merged = pd.merge(df_11, df3, how = 'left', left_on=['camID1'], right_on=['CAMERA ID'])
merged2 = pd.merge(merged, df3, how = 'left', left_on=['camID2'], right_on=['CAMERA ID'])
```

```python
merged2.VIOLATIONS_y.fillna(0, inplace = True) 
merged2['Total_Violations'] = merged2['VIOLATIONS_x'] + merged2['VIOLATIONS_y']
merged2.drop(columns=['camID1','camID2','CAMERA ID_x','VIOLATIONS_x','CAMERA ID_y','VIOLATIONS_y'],inplace=True)
```

```python
merged2.head()
```

```python
merged2.to_csv('CrashandViolations_merged.csv')
```
